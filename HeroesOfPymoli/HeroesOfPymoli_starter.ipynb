```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

```python
total_players = len(purchase_data)
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
## leaving this cell intact to keep model answer
```

```python
average_price = purchase_data['Price'].mean()
unique_items = len(purchase_data['Item ID'].unique())
purchases = len(purchase_data['Purchase ID'].unique())
total_revenue = purchase_data['Price'].sum()

#creating summary DataFrame
summary_df = pd.DataFrame({
    "Number of Unique Items": [unique_items], 
    "Average Price": ["${:.2f}".format(average_price)],
    "Number of Purchases" : [purchases],
    "Total Revenue" : ["${:.2f}".format(total_revenue)]
    })
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
## leaving this cell intact to keep model answer
```

```python
male_count = len(purchase_data[purchase_data['Gender'] == 'Male']['SN'].unique())
female_count = len(purchase_data[purchase_data['Gender'] == 'Female']['SN'].unique())
other_count = len(purchase_data[purchase_data['Gender'] == 'Other / Non-Disclosed']['SN'].unique())
total_players = male_count + female_count + other_count

## Gender Demographics
grouped_gender = purchase_data.groupby(['Gender'])
#aggregate by SN to have the unique values
grouped_gender_unique = grouped_gender.agg({'SN' : 'nunique'})
grouped_gender_unique.rename(columns = {'SN': 'Total Count'}, inplace = True)
percentages = [female_count/ total_players, male_count/total_players, other_count/total_players]


grouped_gender_unique["Percentage of Players"] = percentages
grouped_gender_unique["Percentage of Players"] = grouped_gender_unique["Percentage of Players"].map("{:.2%}%".format)
grouped_gender_unique.sort_values("Total Count", ascending = False)
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
## leaving this cell intact to keep model answer
```

```python
# Purchase count
female_purchase_count = len(purchase_data[purchase_data['Gender'] == "Female"])
male_purchase_count = len(purchase_data[purchase_data['Gender'] == "Male"])
other_purchase_count = len(purchase_data[purchase_data['Gender'] == "Other / Non-Disclosed"])
purchases_count = [female_purchase_count, male_purchase_count, other_purchase_count]

#Other / Non-Disclosed
# Average Purchase Price
female_avg_purchase = purchase_data[purchase_data['Gender'] == "Female"]['Price'].mean()
male_avg_purchase = purchase_data[purchase_data['Gender'] == "Male"]['Price'].mean()
other_avg_purchase = purchase_data[purchase_data['Gender'] == "Other / Non-Disclosed"]['Price'].mean()
avg_purchases = [female_avg_purchase, male_avg_purchase, other_avg_purchase]

# Total Purchase Value
female_total_purchase = purchase_data[purchase_data['Gender'] == "Female"]['Price'].sum()
male_total_purchase = purchase_data[purchase_data['Gender'] == "Male"]['Price'].sum()
other_total_purchase = purchase_data[purchase_data['Gender'] == "Other / Non-Disclosed"]['Price'].sum()
total_purchases = [female_total_purchase, male_total_purchase, other_total_purchase]

# Avg Total Purchase per Person
# Carrying forward from previous calc values for male_count , female_count and other_count
male_avg_total_purchase = male_total_purchase / male_count
female_avg_total_purchase = female_total_purchase / female_count
other_avg_total_purchase = other_total_purchase / other_count
avg_total_purchases_pp = [female_avg_total_purchase, male_avg_total_purchase, other_avg_total_purchase ]

# Creating Data Frame with previous values
purchase_analysis_df = pd.DataFrame({
    "Gender" : ["Females", "Male", "Other / Non-Disclosed"],
    "Purchase Count" : purchases_count,
    "Average Purchase Price": avg_purchases,
    "Total Purchase Value": total_purchases,
    "Avg Total Purchase per Person": avg_total_purchases_pp
})

#changing format
purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].map("${:.2f}".format) 
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_df["Avg Total Purchase per Person"] = purchase_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis_df.set_index(["Gender"])
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
## leaving this cell intact to keep model answer
```

```python
#Defining bins and labels
categories = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+" ]
bins = [0, 9, 14, 19, 24, 29, 34, 39, 49]

# adding a new category to the data frame
purchase_data["Categories"] = pd.cut(purchase_data["Age"], bins, labels = categories, include_lowest = True)

age_groupby_df = purchase_data.groupby(["Categories"])


age_demographics = pd.DataFrame({
    "Total Count":age_groupby_df["SN"].nunique()
    })

#finding the percentage of players
age_demographics["Percentage of Players"] = (age_demographics["Total Count"]/total_players).map("{:.2%}".format)

age_demographics
```

## Top Spenders

* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

```python
# grouping by 'SN'
top_spenders = purchase_data.groupby(['SN'])
# creating data frame
top_spenders_df = pd.DataFrame({
    "Purchase Count" : top_spenders['Purchase ID'].count(),
    "Average Purchase Price" : top_spenders['Price'].mean(),
    "Total Purchase Value" : top_spenders['Price'].sum() 
})

# sorting first to circumbent 'object' type after .map formatting 
top_spenders_df.sort_values("Total Purchase Value", ascending = False, inplace = True)
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format) 
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)
top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
## leaving this cell intact to keep model answer
```

```python
# grouping by Item ID first, then  Item Name
items = purchase_data.groupby(['Item ID','Item Name'])
items_df = pd.DataFrame({
     "Purchase Count" : items['Purchase ID'].count(),
    "Item Price" : items['Price'].mean(),
    "Total Purchase Value" : items['Price'].sum()
})

# sorting first to circumbent 'object' type after .map formatting 
items_df.sort_values("Purchase Count", ascending = False, inplace = True)

# formatting
items_df["Item Price"] = items_df["Item Price"].map("${:.2f}".format) 
items_df["Total Purchase Value"] = items_df["Total Purchase Value"].map("${:.2f}".format)
items_df.head()
```
